# Set membership

```python
myset={"one","two","three","four","five","six"}
myset
```

```python
"one" in myset
```

```python
"ten" in myset
```

# ADD

```python
myset.add("ten")
myset
```

# Update

```python
myset.update(["eleven","twelve","twenty"])
myset
```

# Remove 

```python
myset.remove("ten")
myset
```

-<b> Because twentyone doesnt exist in Upper Set <b>

```python
myset.remove("twentyone")
myset
```

# Discard

```python
myset.discard("twentyone")
myset
```
