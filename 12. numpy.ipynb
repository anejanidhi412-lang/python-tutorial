# Combining the array -- matrices


```python
x=np.array([[12,23],[45,56]])
print(x)
print()
y=np.array([[89,3],[12,76]])
print(y)
print()
#hstack--- horizontal stack
print(np.hstack((x,y)))
```

 <b> hstack is for horizontal stack <b>

 <b>another way <b>

```python
x=np.array([[11],[12],[13]])
print(x)
print()
y=np.array([[14],[15],[16]])
print(y)
print()
print(np.hstack((x,y)))
```

```python
x=np.array([[12,11,13],[14,15,16]])
print(x)
print()
y=np.array([[17,18,19],[1,2,3]])
print(y)
print()
#vstack--- vertical stack
print(np.vstack((x,y)))
```

<b> vstack is for vertical Stack  <b>

```python
#combining two matrices
x=np.array([[12,13],[14,15]])
print(x)
print()
y=np.array([[1,2],[3,4]])\

print(y)
print()
np.concatenate((x,y),axis=1) #hstack
```
