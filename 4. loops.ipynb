# Prime Number

```python
n=int(input("enter the number"))
co=0
i=1
while i<=n:
    if n%1==0:
        co=co+1
    i=i+1
    if co==2:
       print("prime number")
else:
    print("not a prime number")
```

# Reverse of a number

```python
n=int(input("enter the number"))
rev=0
while n>0:
    rem=n%10
    rev=rev*10+rem
    n=n//10
    print("the reverse is",rev)
```
