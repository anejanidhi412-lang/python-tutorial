# Normal Copy

```python
x=[12,34,78,89,67]
y=x
print(x)
print(y)
y[2]="nidhi"
print(x)
print(y)
```

```python
#id
x=[12,34,78,89,67]
y=x
print(x)
print(y)
y[2]="nidhi"
print(x)
print(y)
print(id(x))
print(id(y))
```

# Shallow Copy

```python
x=[12,34,78,89,67]
y=x.copy()
print(x)
print(y)
y[2]="nidhi"
print(x)
print(y)
```
