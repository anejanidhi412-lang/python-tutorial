# String Methods

# 1. Capitalize

```python
s1="hello how are you"
s1.capitalize()
```

# 2. EndsWith

```python
s1="nidhi@gmail.com"
print(s1.endswith("com"))
print(s1.endswith(".com"))
print(s1.endswith("gmail.com"))
print(s1.endswith("@gmail.com"))
print(s1.endswith("co,in"))
```

# 3. Isalnum

```python
s1="1234"
print(s1.isalnum())
s2="abcd"
print(s2.isalnum())
s3="abcd1234"
print(s3.isalnum())
s4="@gmail.com"
print(s4.isalnum())
s5="nidhi aneja"
print(s5.isalnum())
```

# 4. Islower

```python
s1="hello"
print(s1.islower())
s2="hello123"
print(s2.islower())
s3="helllo*&#"
print(s3.islower())
s4="Hello123"
print(s4.islower())
s5="hello how"
print(s5.islower())
s6="Hello how"
print(s6.islower())
```
