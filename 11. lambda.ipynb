# Map

 <b>if you want some changes in your container <b>

```python
list(map(lambda x: x+x,[1,2,3,4]))
```

```python
tuple(map(lambda x:x*x,(1,2,3,4)))
```

```python
my_friends =["radhika","himanshu","parth","akshara"]
list(map(lambda x: x.upper(),my_friends))
```

```python
number1=[1,2,3]
number2=[4,5,6]
number3=[1,5,9]
result = map(lambda x,y,z : x+y+z, number1, number2, number3)
list(result)
```

```python
my_strings =["a","b","c","c"]
my_number =[1,2,3,4,5,6]
my_name=["xx","yy","zz","uu"]
result= list(map(lambda x,y,z :(y,x+z), my_strings,my_number,my_name))
result
```

# Filter

 <b>use to extract data from dataset/container<b>

```python
my_list =[1,2,5,6,7,4,9,8]
new_list = list(filter(lambda x: (x%2==0),my_list))
print(new_list)
```

```python
li=[88,90,56,3,4,5,9]
final_list = list(filter(lambda x: (x%2!=0),li))
print(final_list)
```
