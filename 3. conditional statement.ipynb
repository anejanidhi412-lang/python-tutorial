# 4. Nested If

```python
a=int(input("enter the number"))
b=int(input("enter the number"))
c=int(input("enter the number"))
if a>b:
    if a>c:
        print("A is greatest")
    else:
        print("C is greatest")
else:
    if b>c:
        print("B is greatest")
    else:
        print("C is greatest")
```
